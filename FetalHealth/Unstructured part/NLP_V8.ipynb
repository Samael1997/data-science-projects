```python
import spacy
import numpy as np
from spacy import displacy
import warnings
warnings.filterwarnings("ignore")
```

```python
#CICLARE SUI DOCUMENTI - CREARE OGGETTO PER CIASCUN DOCUMENTO
#DOC LETTURA DA FILE

output = ''
docs = []
for i in range(1,N_DOCS+1):
    with open(f'docs/doc{i}.txt', mode='r', encoding='UTF-8') as d1:
        check = ''
        for line in d1:
            check += line.strip().lower() +" "
        output += check
        docs.append(check)
    #output += '\n'

doc = nlp(output)
```

```python
# TfidfVectorizer 
# CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
import pandas as pd

# Set dei documenti
train = docs

# Istanzio gli oggetti necessari

countvectorizer = CountVectorizer(analyzer='word',stop_words='english',vocabulary = doc2)
tfidfvectorizer = TfidfVectorizer(analyzer='word',stop_words='english',vocabulary = doc2, smooth_idf = True) # UNISCE E OTTIMIZZA IL COUNT VECTORIZER E IL TFIDFTRANSFORMER

# Creazione di due matrici
count_wm = countvectorizer.fit_transform(train)
tfidf_wm = tfidfvectorizer.fit_transform(train)

# Recupero i termini trovati nel corpus e tabellizzo in DataFrames

count_tokens = countvectorizer.get_feature_names()
tfidf_tokens = tfidfvectorizer.get_feature_names()

df_countvect = pd.DataFrame(data = count_wm.toarray(),index = index,columns = count_tokens)
df_tfidfvect = pd.DataFrame(data = tfidf_wm.toarray(),index = index,columns = tfidf_tokens)

print("Count Vectorizer\n")
print(df_countvect)
print("\nTD-IDF Vectorizer\n")
print(df_tfidfvect)
```

```python
surrogate_dict = {}
count_docs = 0

for index, row in df_tfidfvect.iterrows():
    
    list_temp = []
    final_list = []
    
    for (name,value) in row.items():
        # VALORI TF-IDF POSITIVI
        if value>0:
            list_temp.append(name)
    
    
    doc_lenght = len(docs[count_docs])
    for word in list_temp:
        
        #TF CON SCALA LOGARITMICA
        final_list.append((word,1+np.log(docs[count_docs].count(word))))
        
    count_docs += 1
    surrogate_dict[index] = final_list
    
surrogate_dict
```

## Generazione file sql

```python
def createDDL():
    
    with open(f'ddl.sql', mode='w', encoding='UTF-8') as file:
   
    
        # Creo il db

        #db_stm = "CREATE DATABASE medical_text_dB;"
        #file.write(db_stm+"\n")
        #file.write("\n")
        
        table_1_stm =  """
        DROP TABLE IF EXISTS doc;
        CREATE TABLE doc (
        id_doc INTEGER NOT NULL, 
            """
        size = len(df_tfidfvect2.columns)
        for index in df_tfidfvect2.columns:
            
            if size!= 1:
                table_1_stm = table_1_stm + " t_"+ index + " CHARACTER(30) ,"
            else:
                table_1_stm = table_1_stm + " t_" + index + " CHARACTER(30) );"
            size -= 1
        file.write(table_1_stm+"\n")
        alter = """
                ALTER TABLE doc ALTER COLUMN id_doc ADD GENERATED ALWAYS AS IDENTITY (
                START WITH 1
                INCREMENT BY 1
                NO MINVALUE
                NO MAXVALUE
                CACHE 1
            );
            """
        file.write(alter+"\n")
        
        table_2_stm = """
        DROP TABLE IF EXISTS unstructured_doc;
        CREATE TABLE unstructured_doc (
        doc_id INTEGER NOT NULL,
        file_name CHARACTER(10) NOT NULL,
        content CHARACTER(255) NOT NULL);"""
        file.write(table_2_stm+"\n")
        file.write("\n")
        
        alter_2_stm = """
        ALTER TABLE unstructured_doc
        DROP COLUMN content;
        ALTER TABLE unstructured_doc
        ADD COLUMN content CHARACTER(1000);
                """
        
        file.write(alter_2_stm+"\n")
        file.write("\n")
        
        alter_3_stm = """
        ALTER TABLE unstructured_doc ALTER COLUMN doc_id ADD GENERATED ALWAYS AS IDENTITY (
        START WITH 1
        INCREMENT BY 1
        NO MINVALUE
        NO MAXVALUE
        CACHE 1);
            """
        file.write(alter_3_stm+"\n")
        file.write("\n")
        
        alter_4_stm = """
        ALTER TABLE doc
        ADD UNIQUE (id_doc);

        ALTER TABLE unstructured_doc
        ADD UNIQUE (doc_id);

        """
        file.write(alter_4_stm+"\n")
        file.write("\n")
        file.write("\n")
        file.close()
```

```python
createDDL()
```

```python
def createDML():
    
    with open(f'dml.sql', mode='w', encoding='UTF-8') as file:
   
    
        # Inserisco nella tabella

        file.write("TRUNCATE TABLE doc RESTART IDENTITY; \n")
        num_docs = len(surrogate_dict2)
        num_tokens = len(surrogate_dict2[0])
        
        insert_first = "INSERT INTO doc ("
        
        size = len(df_tfidfvect2.columns)
        for index in df_tfidfvect2.columns:
            if size!= 1:
                insert_first = insert_first+"t_" +index + ","
            else:
                insert_first = insert_first+"t_"+ index
            size -= 1
        insert_last = insert_first+ ")VALUES ("
        
        
        for i in range(0,num_docs):
            temp_insert = insert_last
            for j in range(0,num_tokens):
                if surrogate_dict2[i][j][1] < 0:
                    temp_insert = temp_insert + "'0',"
                else:
                    temp_insert = temp_insert + f"'{surrogate_dict2[i][j][1]}',"
            file.write(temp_insert[:-1]+");")
            file.write("\n")
            file.write("\n")
        #print(insert_last)
        file.write("TRUNCATE TABLE unstructured_doc RESTART IDENTITY; \n")
        for i in range(1,N_DOCS+1):
            with open(f'docs/doc{i}.txt', mode='r', encoding='UTF-8') as d1:
                temp = ""
                for line in d1:
                    temp += line.strip().lower().replace("'","")+ " "
                file.write(f"INSERT INTO unstructured_doc (file_name,content) VALUES ('doc{i}.txt','{temp}\');")   
           
        file.write("\n")
        file.close()
```

```python
createDML()
```

```python
def createQL():
    
    with open(f'ql.sql', mode='w', encoding='UTF-8') as file:
   
    
        # Inserisco nella tabella
        num_docs = len(surrogate_dict2)
        num_tokens = len(surrogate_dict2[0])
        
        for tok in new_disease:
            file.write(f"""
            SELECT id_doc,t_{tok}
            FROM doc
            WHERE CAST(t_{tok} as decimal)>0
            ORDER BY t_{tok} DESC;
            """)  
            file.write("\n")
            file.write("\n")
        
        for tok in new_chemical:
            file.write(f"""
            SELECT id_doc,t_{tok}, content
            FROM doc JOIN unstructured_doc on doc.id_doc = unstructured_doc.doc_id
            WHERE CAST(t_{tok} as decimal)>0
            ORDER BY t_{tok} DESC;
            """)  
            file.write("\n")
            file.write("\n")
    
           
        file.write("\n")
        file.close()
        return len(new_chemical)+len(new_disease)
```

```python
print(f"Numero di query generate : ",createQL())
```

```python
# CREO IL DB DEI DOCUMENTI

import psycopg2

from psycopg2 import Error

def connect_and_create_db(database):
    try:
        # Connect to an existing database
        connection = psycopg2.connect(user="postgres",
                                      password="admin",
                                      host="127.0.0.1",
                                      port="5432")

        connection.autocommit = True
        
        # Create a cursor to perform database operations
        cursor = connection.cursor()

        # Executing a SQL query
        cursor.execute(f"CREATE DATABASE {database};")
        
        return (connection,cursor)

    except (Exception, Error) as error:
        # Connect to an existing database
        connection = psycopg2.connect(user="postgres",
                                      password="admin",
                                      host="127.0.0.1",
                                      port="5432",
                                      database=database)
        
        connection.autocommit = True

        # Create a cursor to perform database operations
        cursor = connection.cursor()

        print("PostgreSQL server information")
        print(connection.get_dsn_parameters(), "\n")

        # Executing a SQL query
        cursor.execute("SELECT version();")

        # Fetch result
        record = cursor.fetchone()
        print("You are connected to - ", record, "\n")
        
        return (connection,cursor)
        
def ddl_operation(connection,cursor,table):
     try:

        # Executing a SQL query
        cursor.execute(table)
        
        print("Operation completed!")
    
     except (Exception, Error) as error:
        print("", error)
        
def dml_operation(connection,cursor,insert):
     try:

        # Executing a SQL query
        cursor.execute(insert)
    
     except (Exception, Error) as error: 
        print("", error)
    
def close_connection(connection):
     if connection:
        cursor.close()
        connection.close()
        print("PostgreSQL connection is closed")
        
```
